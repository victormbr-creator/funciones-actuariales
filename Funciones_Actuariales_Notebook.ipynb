## 2. Funciones financieras básicas

La librería también permite calcular factores financieros útiles en matemática actuarial.

```python
i = 0.05

print("Factor de descuento v =", fa.factor_descuento(i))
print("Tasa de descuento d =", fa.tasa_descuento(i))
print("Fuerza de interés delta =", fa.fuerza_interes(i))
print("Valor presente de 1000 a 3 periodos =", fa.valor_presente(1000, i, 3))
print("Valor futuro de 1000 a 3 periodos =", fa.valor_futuro(1000, i, 3))
```

## 3. Anualidades actuariales

Ahora calculamos anualidades para una vida de edad 0 con tasa de interés del 5%.

```python
edad = 21
n = 10

a_temp_ant = fa.anualidad_temporal_anticipada(edad, n, i, tabla)
a_temp_inm = fa.anualidad_temporal_inmediata(edad, n, i, tabla)
a_vit_ant = fa.anualidad_vitalicia_anticipada(edad, i, tabla)

print("Anualidad temporal anticipada =", a_temp_ant)
print("Anualidad temporal inmediata =", a_temp_inm)
print("Anualidad vitalicia anticipada =", a_vit_ant)
```

## 4. Seguros actuariales

También es posible calcular distintos tipos de seguros discretos.

```python
seg_temp = fa.seguro_temporal(edad, n, i, tabla)
seg_dotal = fa.seguro_dotal(edad, n, i, tabla)

print("Seguro temporal =", seg_temp)
print("Seguro dotal =", seg_dotal)
```

## 5. Funciones de conmutación

Las funciones de conmutación resumen información actuarial útil para cálculos más compactos.

```python
print("D_0 =", fa.Dx(0, i, tabla))
print("N_0 =", fa.Nx(0, i, tabla))
print("C_0 =", fa.Cx(0, i, tabla))
print("M_0 =", fa.Mx(0, i, tabla))
print("R_0 =", fa.Rx(0, i, tabla))
```

## 6. Valor presente de flujos

La librería también incluye funciones generales para descontar flujos de efectivo.

```python
flujos = [100, 200, 300, 400]

vp_flujos = fa.valor_presente_flujos(flujos, i)
vp_monto = fa.valor_presente_monto(1000, i, 4)

print("Valor presente de flujos =", vp_flujos)
print("Valor presente de un monto único =", vp_monto)
```

## 7. Mini caso de uso

Supongamos que queremos comparar el valor actuarial de:

- una anualidad temporal anticipada de 3 años
- un seguro temporal de 3 años

para una vida de edad 0 y una tasa del 5%.

```python
edad = 0
tasa = 0.05
n = 3

anualidad = fa.anualidad_temporal_anticipada(edad, n, tasa, tabla)
seguro = fa.seguro_temporal(edad, n, tasa, tabla)

print("Anualidad temporal anticipada:", anualidad)
print("Seguro temporal:", seguro)
print("Diferencia:", anualidad - seguro)
```
